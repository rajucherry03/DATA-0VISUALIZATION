```python


import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (you may have your actual data)
data = {
    'Country': ['India', 'France'],
    'Year 2000': [1000, 2000],
    'Year 2001': [1200, 2100],
    # Add more years as needed
}
df = pd.DataFrame(data)
df.set_index('Country', inplace=True)
fnew4 = df.loc[['India', 'France'], :]
fig1, ax1 = plt.subplots()
total_immigrants = fnew4.sum()
labels = total_immigrants.index
sizes = total_immigrants
explode = (0.1, 0)
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Title for the Pie Chart
plt.title('Total Immigrants by Country (Pie Chart)')

# Display the Pie Chart
plt.show()

```

```python
#Pie chart:
fig1,axl=plt.subplots()
fnew4=pd.DataFrame()
for i,j in zip([0,11], ['India','France']):
  fnew4[j]=df.iloc[i,4]
l4=['India','France']
pnew4=fnew4.transpose()
axl.pie(pnew4['Total'],labels=l4)
ax1.axis('equal')
plt.title('Total Immigrants by Country (Pie Chart)')
plt.show()
```
